# Extracting meta data and exporting as CSV
    

```python
download_dir = "Meta_Data_Positive.csv" #where you want the file to be downloaded to 
csv = open(download_dir, "w") 
#"w" indicates that you're writing strings to the file
columnTitleRow = "Video, frames, width, height, fps\n"
csv.write(columnTitleRow)

path = 'D:/Workspace/__Workstation/4th year project/CrashCatcher-master/My work/Positive/' ### specify the path
files = os.listdir(path)

for file in files:

    cap = cv2.VideoCapture(path+file)

    if not (cap.isOpened()): 
        print ("could not open :"+file)
        
    #frames = str(cap.get(7))
    
    frames = str(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width  = str(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = str(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = str(cap.get(cv2.CAP_PROP_FPS))

    row = file + "," + frames +"," + width +"," + height + "," + fps + "\n"
    print (row)
    #uncomment the below line
    csv.write(row)
```

# Chopping extra frames to make Max frames as 120 or less

```python
from moviepy.editor import VideoFileClip


long=['NoAccident_014.mp4']

for x in long:
        clip = VideoFileClip(x) 
        newclip = clip.subclip(0,4)
        ### CAUTION this function gets 4 sec video only if it longer than 4 .. running it multiple times may give you just frames
        newclip.write_videofile(x,audio=False) # don't render audio.  
        print(x)

```
